# Median Age of Death

The second metric is based on each neighborhood's median age of death or life expectancy. The dataset I used was https://data.wprdc.org/dataset/median-age-death. Median age of death is an indicator of the health standard in different areas of Pittsburgh.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
pgh_md_death = pd.read_csv("https://data.wprdc.org/dataset/ee735209-4de7-4ea4-b446-bf0f0f6d9cb3/resource/c2e1500a-a12a-4e91-be94-76c6a892b7e2/download/nhoodmedianage20112015.csv",
                          usecols = ['NEIGHBORHOOD','TOTAL MD AGE AT DEATH'])

pgh_md_death=pgh_md_death.dropna()

pgh_md_death = pgh_md_death.sort_values(by='TOTAL MD AGE AT DEATH', ascending=False)
pgh_md_death = pgh_md_death.head(30)

pgh_md_death.plot.bar(x = 'NEIGHBORHOOD',title = "Median age of Death", figsize=(10,5))
```

# Second Metric Conlusion

In conclusion the best neighborhood by this metric is Squirrel Hill South. This has the highest median age of death which is an indicator of a healthier standard of living and prosperity. Some other close neighborhoods are North Oakland, Banksville, Point Breeze, and Squirrel Hill North.
